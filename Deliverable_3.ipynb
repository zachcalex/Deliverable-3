```python
#Importing libraries
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
#Loading Iris dataset
iris = sns.load_dataset('iris')
iris.head()
```

```python
#Dropping the "species" column
iris = iris.drop(labels = 'species', axis = 1)
iris.head()
```

```python
#Adding columns
iris['sepal_sum'] = iris['sepal_length'] + iris['sepal_width']
iris.head()
```

```python
#Create new data frame with summary stats
summary_stats = pd.DataFrame({
    'mean': iris.mean(),
    'std': iris.std(),
    'min': iris.min(),
    'max': iris.max()
})
print(summary_stats)
```
